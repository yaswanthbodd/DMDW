```python
import numpy as np
from numpy.linalg import norm
 
# define two lists or array
A = np.array([1,1,1,1,1,1,0,0,0])
B = np.array([0,0,1,0,1,1,1,1,1])

# compute cosine similarity8
cosine = np.dot(A,B)/(norm(A)*norm(B)) #||A|| is L2 norm of A: It is computed as square root of the sum of squares of elements of the vector A.
print("Cosine Similarity:", cosine)
```

```python
#define Jaccard Similarity function
def jaccard(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union

a = [0, 1, 2, 5, 6, 8, 9]
b = [0, 2, 3, 4, 5, 7, 9]
jaccard(a, b)
```
