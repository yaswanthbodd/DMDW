```python
# 8.k-means clustring
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
ratings=[['yash',5,4,2,1],
        ['subbu',3,4,2,1],
        ['sunil',4,4,1,2],
        ['srinu',1,2,4,4],
        ['pawan',1,2,5,4],
        ['syam',1,2,5,5]]
titles=['user','rrr','kgf','jathi','f2']
movies=pd.DataFrame(ratings,columns=titles)
movies
```

```python
from sklearn import cluster
data=movies.drop('user',axis=1)
data
k_means=cluster.KMeans(n_clusters=2,max_iter=10,random_state=0)
k_means.fit(data)
```

```python
labels=k_means.labels_
pd.DataFrame(labels,index=movies.user,columns=['Cluster ID'])
```

```python
centroids=k_means.cluster_centers_
pd.DataFrame(centroids,columns=data.columns)
```

```python
testdata=np.array([[4,5,1,2],
                  [5,4,1,3],
                  [4,4,2,3],
                  [1,3,5,5],
                  [5,4,1,4]])
labels=k_means.predict(testdata)
labels=labels.reshape(-1,1)
usernames=np.array(['sunny','sai','sankar','dhoni','yuvaraj']).reshape(-1,1)
cols=movies.columns.tolist()
cols.append('Cluster ID')
newuser=pd.DataFrame(np.concatenate((usernames,testdata,labels),axis=1),columns=cols)
newuser
```

```python
plt.scatter(centroids[:,0],centroids[:,1])
plt.xlabel('Ratings')
plt.ylabel('Movies')
```

```python
%matplotlib inline 
numclusters=[1,2,3,4,5,6]
SSE=[]
for k in numclusters:
    k_means=cluster.KMeans(n_clusters=k)
    k_means.fit(data)
    SSE.append(k_means.inertia_)
plt.plot(numclusters,SSE)
plt.xlabel('number of clusters')
plt.ylabel('standard square root')
```

```python
#10. dbscan algorithm
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import DBSCAN
from scipy.spatial.distance import cdist
```

```python
y = np.array([[5, 7], [8, 4], [3, 3], [4, 4], [3, 7], [6, 7], [6, 1], [5, 5]])
dist_matrix=cdist(y,y,metric='euclidean')
dist_matrix
df=pd.DataFrame(dist_matrix)
df
```

```python
eps=3.5
min_samples=3
dbscan=DBSCAN(eps=eps,min_samples=min_samples,metric='euclidean')
dbscan
labels=dbscan.fit_predict(y)
labels
```

```python
core_samples_mask=np.zeros_like(labels,dtype=bool)
core_samples_mask[dbscan.core_sample_indices_]=True
```

```python
core_points=y[core_samples_mask]
border_points=y[np.logical_and(~core_samples_mask,labels!=-1)]
noise_points=y[labels==-1]
```

```python
print("Core Points: ")
print("Data Point\tCore Points")
for i,x in enumerate(y):
    core_labels=[j for j,core in enumerate(core_points) if np.array_equal(core,x)]
    core_labels_str=','.join([f's{j+1}'for j in core_labels])
    print(f's{i+1}\t\t{core_labels_str}')
print()
```

```python
plt.figure(figsize=(8,6))
colors=['blue','red','green','purple']
markers=['o','s','^']
for label in set(labels):
    if label ==-1:
        plt.scatter(noise_points[:,0],noise_points[:,1],color='gray',marker='x',label='Noise')
    elif label ==0:
        plt.scatter(core_points[:,0],core_points[:,1],color=colors[label],marker=markers[0],label=f'cluster{label}')
    else:
        cluster_points=y[labels==label]
        plt.scatter(cluster_points[:,0],cluster_points[:,1],color=colors[label%len(colors)],marker=markers[label%len(markers)],label=f'cluster{label}')
if len(border_points)>0:
    plt.scatter(border_points[:,0],border_points[:,1],color='yellow',marker='o',edgecolors='black',linewidths=1,label='Border')

plt.xlabel('x')
plt.ylabel('y')
plt.title('DBSCAN CLustring')
plt.legend()
plt.show()
```
